# Auto Random state

```python
sclist=[]
for i in range(500):
     Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.20,random_state=i)
     model = LinearSVC()
     model.fit(Xtrain,ytrain)
     sc=model.score(Xtest,ytest)
     sclist.append(sc)
m=max(sclist)
print(m)
id=sclist.index(m)
print(id)
```

# Cross validation

```python
from sklearn.model_selection import cross_val_score,KFold,StratifiedKFold
```
